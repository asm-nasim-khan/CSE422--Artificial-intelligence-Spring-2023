```python
from queue import PriorityQueue
import math
from google.colab import drive
drive.mount('/content/drive')
handler = open('/content/drive/MyDrive/CSE422/Lab_1/Input_file.txt')
lines = handler.read().split("\n")
the_graph = {}
the_heu = {}
for line in lines:
  extract = line.split()
  key = extract[0]
  h = int(extract[1])
  if key not in the_graph.values():
    the_graph[key] = []
    the_heu[key] = h

  for idx in range(2,len(extract)):
    if idx % 2 == 0:
      the_graph[key].append(extract[idx])
    else:
      the_graph[key].append(int(extract[idx]))

# print(the_graph)
# print(the_heu)


def A_search(Graph, start, finish,the_heu):
  queue = PriorityQueue()
  visited=[]
  distance= {}
  distance_h= {}
  parent = {}
  for node in Graph.keys():
    distance[node] = math.inf
    distance_h[node] = math.inf
    parent[node] = None
  distance[start] = 0
  distance_h[start] = the_heu[start]
  # print((the_heu[start],start))
  queue.put((the_heu[start],start))
  
  # print(queue.get())
  while not queue.empty():
    Curr_hue,Curr_node = queue.get()
    visited+= Curr_node
  
    for index in range(len(Graph[Curr_node])):
      
      if index%2 == 0:
        # print(Graph[Curr_node][index],Graph[Curr_node][index+1])
        if distance[Graph[Curr_node][index]] > (distance[Curr_node] + Graph[Curr_node][index+1] ):
          distance[Graph[Curr_node][index]] = distance[Curr_node] + Graph[Curr_node][index+1]
          distance_h[Graph[Curr_node][index]] = distance[Graph[Curr_node][index]] + the_heu[Graph[Curr_node][index]] # here
          parent[Graph[Curr_node][index]] = Curr_node
          queue.put((distance[Graph[Curr_node][index]],Graph[Curr_node][index]))
      if Curr_node == finish:
        path= []
        path.append(finish)
        check = finish
        while check!= start:
          check=parent[path[-1]]
          path.append(parent[path[-1]])
        return path,distance[Curr_node]
#=========================
goal = "Bucharest"
pathWay,cost = A_search(the_graph,"Arad",goal,the_heu)
for i in range(len(pathWay)-1,-1,-1):
  if pathWay[i] == goal:
    print(pathWay[i])
  else:
    print(pathWay[i],end=" -> ")
print(f'Total distance: {cost} km')
```
