```python
import math
import random
ID = input("Whats Your ID: ")
alpha = - math.inf
beta = math.inf
def alpha_beta_pruning(depth,max_min,alpha,beta,terminal,position):
  if depth == 0:
    return terminal[position]
  if max_min:
    maxEvaluation = - math.inf
    for i in range(2):
      Evaluation=alpha_beta_pruning(depth-1,False,alpha,beta,terminal,(position*2)+i)
      maxEvaluation=max(maxEvaluation,Evaluation)
      alpha=max(alpha,maxEvaluation)
      if alpha >= beta:
        break
    return maxEvaluation
  else:
    
    minEvaluation = math.inf
    for i in range(2):
      Evaluation=alpha_beta_pruning(depth-1,True,alpha,beta,terminal,(position*2)+i)
      minEvaluation=min(minEvaluation,Evaluation)
      beta=min(beta,minEvaluation)
      if alpha >= beta:
        break
    return minEvaluation

#============================================
ID = ID.replace('0','8')
minimum_point = int(ID[4])
win_point = int(ID[-1:-3:-1])
maximum_point = math.ceil(win_point * 1.5)
print(win_point)
terminal_nodes = [None] * 8
terminal_nodes_temp = [random.randrange(minimum_point,maximum_point) for x in range(8)]
terminal_nodes.extend(terminal_nodes_temp)
print(terminal_nodes)
ans=alpha_beta_pruning(3,True,alpha,beta,terminal_nodes,1)

print(f'Generated 8 random points between the minimum and maximum point: {terminal_nodes_temp}')
print(f'Total points to win: {win_point}')
print(f'Achieved point by applying alpha-beta pruning: {ans}')
if ans>=win_point:
  print('The winner is Optimus Prime')
else:
  print('The winner is Megatron')
```

```python
#task_2
result = []
shuffle = int(ID[3])
# shuffle = 6
# print(shuffle)

for i in range(shuffle):
  random.shuffle(terminal_nodes_temp)
  terminal_nodes_tm =[None] * 8
  terminal_nodes_tm.extend(terminal_nodes_temp)
  result.append(alpha_beta_pruning(3,True,alpha,beta,terminal_nodes_tm,1))

count = sum(i >= win_point for i in result)

print('After the shuffle:')
print(f'List of all points values from each shuffle:{result}')
print(f'The maximum value of all shuffles: {max(result)}' )

print(f'Won {count} times out of {shuffle} number of shuffles')
```
